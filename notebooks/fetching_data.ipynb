```python
from pylossmap.loader import LossMapFetcher
import pandas as pd
```

## Fetching data:

```python
fetcher = LossMapFetcher()
```

### from timestamps:

```python
t1 = pd.to_datetime('2018-10-01 00:00:00').tz_localize('Europe/Zurich')
t2 = pd.to_datetime('2018-10-01 01:00:00').tz_localize('Europe/Zurich')
```

```python
BLM_data = fetcher.from_datetimes(t1 ,t2)
```

```python
BLM_data.data
```

```python
BLM_data.meta
```

```python
BLM_data.plot(min_max_quantile=0.95)
```

### from fill & beam mode:

```python
BLM_data = fetcher.from_fill(7427, beam_modes=['RAMP', 'STABLE', 'INJPHYS'])
```

```python
BLM_data.data
```

```python
BLM_data.meta
```

```python
BLM_data.plot(data=BLM_data.data.loc['RAMP'])
```

```python
BLM_data.plot(data=BLM_data.data.loc['STABLE'])
```

## Finding max values of BLM subset:

```python
BLM_data.find_max()
```

```python
for idx, blms in BLM_data.iter_maxes():
    print(idx)
    print(blms)
    display(BLM_data.data.loc[idx])
```
