```python
from pylossmap.loader import LossMapFetcher
from pylossmap.data import BLMDataCycle
import pandas as pd
```

# get the blm data

```python
fetcher = LossMapFetcher()
```

```python
BLM_data = fetcher.from_fill(7247, beam_modes=['INJPHYS'], unique_beam_modes=True)
# if a fill has multiple of the same beam mode, they will be made unique, use unique_beam_modes=True,
# to control which of them to fetch. To get the second occurance of INJPHYS, use INJPHYS_2, with
# unique_beam_mode=True.
```

```python
BLM_data.data
```

```python
BLM_data.plot()
```

BLM_max, list of BLMs for which to look for the maximum values.

```python
BLM_data.BLM_max
```

"Loss maps" for the timestamps where the BLMs in BLM_max are maximum.

```python
for idx, blms in BLM_data.iter_maxes():
    row = BLM_data.data.loc[idx]
    # will fetch background signal from the first INJPROT beam mode of the fill.
    loss_map = BLM_data.loss_map(row=row,
                                 t=idx[1],
                                 background='fetch',
                                 context=blms)
    loss_map.clean_bg().normalize().plot(title=f"{idx[1]} {idx[0]}\n{'  '.join(blms)}")
```

# processing the LossMap

Plotting the background signal

```python
from pylossmap.plotting import plot_waterfall
```

```python
plot_waterfall(loss_map.background, loss_map.meta, min_max_quantile=0.95)
```

```python
# loss_map.background.mean()
loss_map.plot(data=loss_map.background.mean(), title='background signal')
```

# cleaning and normalizing

```python
LM_clean = loss_map.clean_bg().normalize()
LM_clean.plot()
```

# Filtering BLMs

The LossMap class has some filtering methods, some examples of what can be done is shown below.

```python
LM_clean.beam(1).IR(7).plot()
```

```python
LM_clean.IR(7).beam(2).plot()
```

Dispersion suppressors around IR7

```python
LM_clean.IR(7).DS().plot()
```

TCPs right side of IR7

```python
LM_clean.IR(7).side('R').TCP().plot()
```

Cold collimators in Cells 4,5,6,7,8,9 of IR 5 and 7

```python
LM_clean.cell(4,5,6,7,8,9).type('cold').IR(7,5).plot()
```

```python
LM_clean.type('coll').plot()
```

custom filter

```python
LM_clean.filter(r'BLM(EL|QI)').IR(7).plot()
```

The LossMap's get & set items are forwarded to LossMap.data which allows for easier pandas style filtering

```python
LM_copy = LM_clean.copy()
LM_copy.data = LM_clean[LM_clean['dcum'] < 100000]
LM_copy.plot()
```
