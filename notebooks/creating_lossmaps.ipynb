```python
from pylossmap.loader import BLMDataFetcher
```

# get the blm data

```python
fetcher = BLMDataFetcher()
```

```python
BLM_data = fetcher.from_fill(7247, beam_modes=['INJPHYS'], unique_beam_modes=True)
# if a fill has multiple of the same beam mode, they will be made unique, use unique_beam_modes=True,
# to control which of them to fetch. To get the second occurance of INJPHYS, use INJPHYS_2, with
# unique_beam_mode=True.
```

```python
BLM_data.data
```

```python
BLM_data.plot()
```

Background signal: INJPROT beam mode --> no beam 

```python
BLM_bg = fetcher.bg_from_INJPROT(7247)
BLM_bg.get_beam_meta('intensity', beam=1).plot()
# no intensity --> no beam
BLM_bg.plot()
LM_bg = BLM_bg.loss_map(row=BLM_bg.data.mean())
LM_bg.plot()
```

"Loss maps" for the timestamps where the primary BLMs are maximum.

```python
for idx, blms in BLM_data.iter_max():
    print(idx)
    LM = BLM_data.loss_map(datetime=idx[1],
                           context=blms)
    # the context kwargs allow you to pass some additional information of the loss map's origin
    LM.set_background(LM_bg)
    LM.clean_background().normalize().plot(title=f"{idx[1]} {idx[0]}\n{'  '.join(blms)}")
```

# cleaning and normalizing

```python
LM_clean = LM.clean_background().normalize()
LM_clean.plot()
```

# Filtering BLMs

The LossMap class has some filtering methods, some examples of what can be done is shown below.

```python
LM_clean.beam(1).IR(7).plot()
```

```python
LM_clean.IR(7).beam(2).plot()
```

Dispersion suppressors around IR7

```python
LM_clean.IR(7).DS().plot()
```

TCPs right side of IR7

```python
LM_clean.IR(7).side('R').TCP().plot()
```

Cold collimators in Cells 4,5,6,7,8,9 of IR 5 and 7

```python
LM_clean.cell(4,5,6,7,8,9).type('cold').IR(7,5).plot()
```

```python
LM_clean.type('coll').plot()
```

custom filter

```python
LM_clean.filter(r'BLM(EL|QI)').IR(7).plot()
```

The LossMap's get & set items are forwarded to LossMap.data which allows for easier pandas style filtering

```python
LM_2 = LM_clean[LM_clean.data['dcum'] < 100000]
LM_2.plot()
```
